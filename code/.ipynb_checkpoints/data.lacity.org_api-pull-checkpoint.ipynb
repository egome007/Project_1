* Create a summary data frame to hold the results

```python
#Top Spenders
SN_group = purchase_data.groupby(["SN"])
SN_group.head()

#Calculate the total purchase count by User
purchase_count = SN_group.count()["Price"].rename("Purchase Count")
purchase_count.head()

#Calculate the average purchase price of Items purchase by User
average_purchase_price = SN_group.mean()["Price"].rename("Average Purchase Price")
average_purchase_price.head()

#Calculate the total of all purchases by User
total_purchase_user = SN_group.sum()["Price"].rename("Total Purchase Value")
total_purchase_user.head()

#Place all of the data found into a summary DataFrame
top_spender_summary_table = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": average_purchase_price,
                              "Total Purchase Value": total_purchase_user})
    
top_spender_summary_table
```
